```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = x * x
plt.title("Sine Waveform")
plt.plot(x, y, color='b')
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
sales_one = [160, 150, 140, 145, 175, 165, 180]
sales_two = [70, 90, 160, 150, 140, 145, 175]
plt.figure(figsize=(6,5), dpi=150)
plt.plot(days, sales_one)
plt.plot(days, sales_two)

plt.xlabel('days of the week')
plt.ylabel('sales')
plt.grid()
plt.show()
```
